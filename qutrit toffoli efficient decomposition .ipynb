```python
def twoControlled(q0,q1,q2):   
    q1.CU_1(np.array([[0, 1, 0], [0, 0, 1], [1, 0, 0]]), q0, q1)
    q2.CU_2(np.array([[0, 1, 0], [0, 0, 1], [1, 0, 0]]), q1, q2)
    q1.CU_1(np.array([[0, 0, 1], [1, 0, 0], [0, 1, 0]]), q0, q1)
    return q0
```

```python

n=8

qc = Qutrit(n)    
#initialize qutrits here
for i in range(n-1):
    qc.XP1(i)
#qc.XM1(7)

# circuit for q0, q1, q2 qutrits     
qc.CU_1(np.array([[0, 1, 0], [0, 0, 1], [1, 0, 0]]), 0, 2)
qc.CU_2(np.array([[0, 1, 0], [0, 0, 1], [1, 0, 0]]), 2, 1)
qc.CU_1(np.array([[0, 0, 1], [1, 0, 0], [0, 1, 0]]), 0, 2)

# circuit for q4, q5, q6 qutrits     
qc.CU_1(np.array([[0, 1, 0], [0, 0, 1], [1, 0, 0]]), 4, 6)
qc.CU_2(np.array([[0, 1, 0], [0, 0, 1], [1, 0, 0]]), 6, 5)
qc.CU_1(np.array([[0, 0, 1], [1, 0, 0], [0, 1, 0]]), 4, 6)

# circuit for q1, q3, q5 qutrits     
qc.CU_2(np.array([[0, 0, 1], [1, 0, 0], [0, 1, 0]]), 1, 5)
qc.CU_1(np.array([[0, 1, 0], [0, 0, 1], [1, 0, 0]]), 5, 3)
qc.CU_2(np.array([[0, 1, 0], [0, 0, 1], [1, 0, 0]]), 1, 5)

#q3 is control qutrit and q7 is target qutrit for generalized toffoli
qc.CU_2(np.array([[0, 1, 0], [0, 0, 1], [1, 0, 0]]), 3, 7)

#mirror circuit

# mirror circuit for q1, q3, q5 qutrits      
qc.CU_2(np.array([[0, 0, 1], [1, 0, 0], [0, 1, 0]]), 1, 5)
qc.CU_1(np.array([[0, 0, 1], [1, 0, 0], [0, 1, 0]]), 5, 3)
qc.CU_2(np.array([[0, 1, 0], [0, 0, 1], [1, 0, 0]]), 1, 5)

# mirror circuit for q0, q1, q2 qutrits     
qc.CU_1(np.array([[0, 1, 0], [0, 0, 1], [1, 0, 0]]), 0, 2)
qc.CU_2(np.array([[0, 0, 1], [1, 0, 0], [0, 1, 0]]), 2, 1)
qc.CU_1(np.array([[0, 0, 1], [1, 0, 0], [0, 1, 0]]), 0, 2)

# mirror circuit for q4, q5, q6 qutrits     
qc.CU_1(np.array([[0, 1, 0], [0, 0, 1], [1, 0, 0]]), 4, 6)
qc.CU_2(np.array([[0, 0, 1], [1, 0, 0], [0, 1, 0]]), 6, 5)
qc.CU_1(np.array([[0, 0, 1], [1, 0, 0], [0, 1, 0]]), 4, 6)

qc.get_probs()
```
